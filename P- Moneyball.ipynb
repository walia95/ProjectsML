### FEATURE ENGINEERING
In feature engineering we add more metrics which can help us to better understand the data.
Here we will be adding three more metrics such such as batting average, on base percentage and sluggish percentage.
The above mentioned metrics will need a formula which can be obtained from the formulas on wikipedia.

```python
df['ba']= df['H']/df['AB']
```

```python
df['ba'].tail()
```

```python
# here we willbe calculating OBP.(On base percentage)
df['OBP'] = (df['H']+df['BB']+df['HBP'])/(df['AB']+df['BB']+df['HBP']+df['SF'])
```

```python
df['OBP'].tail()
```

```python
df['1b']= df['H']-df['2B']- df['3B']- df['HR']  #calculated the singles values.
```

```python
df['1b'].tail()
```
